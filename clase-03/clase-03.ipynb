```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20, 200)  # Create a list of evenly-spaced numbers over the range
plt.plot(x, np.sin(x))       # Plot the sine of each x point
plt.show()                   # Display the plot
```

```python
import matplotlib.pyplot as plt
import numpy as np
# senoidales 
tiempo = np.arange(0, 155 ,1)
N = 256

ciclos = 1
amplitud = 1 
seno1 = amplitud * np.sin( ciclos * 2 * np.pi *tiempo/N)
plt.plot(tiempo, seno1) 
plt.show()

print(tiempo)
```
